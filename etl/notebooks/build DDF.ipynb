```python
all_files = [x for x in os.listdir(proc_dir) if 'csv' in x]
```

```python
def pushrows(df, col, match, steps=None, from_col=None):
    if not from_col:
        from_col = '0'
        
    df_ = df.copy()
    
    for i, row in df_[df_[col].str.match(match).fillna(False)].iterrows():
        r = row[int(from_col):]
        if steps:
            for _ in range(steps):
                df_.ix[i, from_col:] = r.shift(1)
        else:
            for _ in range(len(row)):
                if pd.isnull(r[-1]):
                    r = r.shift(1)
                else:
                    df_.ix[i, from_col:] = r
                
    return df_
```
