```python
def format_table(df):
    df = df.ix[3:]
    for c in df.columns:
        if df[c].dropna().empty:
            df = df.drop(c, axis=1)
    if len(df.columns) == 2:
        df.columns = ['indicator', 'value']
    elif len(df.columns) == 3:
        df.columns = ['indicator', 'description', 'value']
    elif len(df.columns) == 4:
        df.columns = ['mics_id', 'indicator', 'description', 'value']
    else:
        raise ValueError(df.columns)
    return df
```

```python
errors = []

for i, f in enumerate(fs):
    try:
        df = pd.read_csv(StringIO('\n'.join(f)))
    except:
        errors.append(f)
        df = pd.read_csv(StringIO('\n'.join(f)), error_bad_lines=False, warn_bad_lines=True, engine='c')
    name = df.columns[0]
    if name.startswith('Unname'):
        name = 'unname_{}'.format(i)
    df = format_table(df)
    df.to_csv("./output/Belize/table_{}.csv".format(to_concept_id(name)), index=False)
```
